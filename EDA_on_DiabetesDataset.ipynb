*Why EDA?*

- helps in getting the clear view of what dataset actually contain.

## Types of EDA

## ` Univarient analysis `
- focus on analyzing one column at a time
- helps in understanding the distribution , outlier , statistical summary
- Technique included (histogram , boxplot , kde , mean , median etc)

## `Bivarient analysis`
- helps in indentifying the relation between two features
- helps in understanding correaltion and dependencies
- Techniques (scatter plot , correlation matrix , barplot , Etc)

## `Multivarient analysis`
- Analyze interaction between more than two feature
- Useful for feature extraction and pattern recognition
- Techniques: Pair plots, PCA (Principal Component Analysis), Heatmaps.

# Bivarient analysis

Bivariate analysis is the process of examining the relationship between two variables in a dataset. It helps in understanding how one variable influences or correlates with another. This is useful for identifying trends, dependencies, or patterns between features.

## Types of Bivariate Analysis & Techniques
Bivariate analysis is divided into three categories based on the type of data:

`Numerical vs. Numerical`

Used when both variables are continuous (numeric).

*Techniques:*

*Scatter Plot:* Visualizes the relationship between two numerical variables.

*Correlation Coefficient (Pearson/Spearman):* Measures the strength and direction of the relationship.

`Categorical vs. Numerical`

Used when one variable is categorical and the other is numerical.

*Techniques:*

*Box Plot:* Compares distributions of a numerical variable across different categories.
*Violin Plot:* Similar to a box plot but shows density distribution.
Bar Chart: Shows the mean/median of a numerical variable for each category.

`Categorical vs. Categorical`


Used when both variables are categorical.

Techniques:

Chi-Square Test: Checks for dependence between two categorical variables.
Stacked Bar Chart: Compares proportions within categories.

Heatmap: Uses color intensity to represent frequency or correlation between categories.

```python


# Grouping data by 'Pregnancies' and calculating the mean Outcome (proportion of diabetics)
pregnancy_diabetes = data.groupby('Pregnancies')['Outcome'].mean()

# Plotting a line graph
plt.figure(figsize=(8,5))
plt.plot(pregnancy_diabetes.index, pregnancy_diabetes.values, marker='o', linestyle='-', color='b')

# Labels and title
plt.xlabel('Number of Pregnancies')
plt.ylabel('Proportion of Diabetes Cases')
plt.title('Diabetes Risk vs Number of Pregnancies')
plt.grid(True)

plt.show()
```
